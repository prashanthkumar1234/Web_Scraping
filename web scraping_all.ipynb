```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
phn_name=[]
phn_price=[]
phn_org=[]
page_num=input('enter the number of pages :')
for i in range(1,int(page_num)+1):
    url="https://www.flipkart.com/search?q=laptops&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_4_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_4_0_na_na_na&as-pos=4&as-type=TRENDING&suggestionId=laptops&requestId=94e4adfd-4834-4e77-8b39-d66b4f6c35ef&page="+str(i)
    req=requests.get(url)
    content=BeautifulSoup(req.content,'html.parser')
    name=content.find_all('div',{'class':'_4rR01T'})
    price=content.find_all('div',{'class':'_30jeq3 _1_WHN1'})
    original_price=content.find_all('div',{'class':'_3I9_wc _27UcVY'})
    print("phone in page" +" "+ str(i))
    
    
    for i in name:
        phn_name.append(i.text)
    for i in price:
        phn_price.append(i.text)
    for i in original_price:
        phn_org.append(i.text)
    
```

```python
print(len(phn_name))
print(len(phn_price))
print(len(phn_org))
```

```python
for i in phn_name:
    print(i)
for i in phn_price:
    print(i)
for i in phn_org:
    print(i)
```

```python
data={"phone name":phn_name,"phone price":phn_price,"original_price":phn_org}
df=pd.DataFrame(data)
print(df)
```
